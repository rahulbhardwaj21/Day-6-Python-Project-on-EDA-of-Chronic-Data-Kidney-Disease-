```python
#Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#Read CSV File using Pandas
df = pd.read_csv('/content/kidney_disease.csv')
```

#**Profile of Data (Basic Analysis of Data)**

```python
df.head()
```

```python
df.columns
```

```python
df.shape
```

```python
#Cleaning of Data
#Drop ID because it is same as Index
df.drop('id', axis = 1, inplace = True)
```

```python
df.head()
```

```python
#Change short form of column header
df.columns = ['age', 'blood_pressure', 'specific_gravity', 'albumin', 'sugar', 'red_blood_cells', 'pus_cell',
              'pus_cell_clumps', 'bacteria', 'blood_glucose_random', 'blood_urea', 'serum_creatinine', 'sodium',
              'potassium', 'haemoglobin', 'packed_cell_volume', 'white_blood_cell_count', 'red_blood_cell_count',
              'hypertension', 'diabetes_mellitus', 'coronary_artery_disease', 'appetite', 'peda_edema',
              'aanemia', 'class']
```

```python
df.head()
```

```python
#Statistic Data
df.describe()
```

```python
#Check null values
df['packed_cell_volume'].isna().sum()
```

```python
#Convert Data Type using Pandas because astype method create value error.
#Use coerce to suppress the error.
df['packed_cell_volume'] = pd.to_numeric(df['packed_cell_volume'], errors = 'coerce')
df['white_blood_cell_count'] = pd.to_numeric(df['white_blood_cell_count'], errors = 'coerce')
df['red_blood_cell_count'] = pd.to_numeric(df['red_blood_cell_count'], errors = 'coerce')
```

```python
df.info()
```

```python
#Categorical and Numerical Columns
cat_cols = [col for col in df.columns if df[col].dtypes == "object"]
num_cols = [col for col in df.columns if df[col].dtypes != "object"]
```

```python
cat_cols
```

```python
num_cols
```

```python
#Create a function which gives value of all categorical columns.
for col in cat_cols:
  print(f"{col} has {df[col].unique()} values \n")
```

```python
#Replacing Unwanted Values
df['diabetes_mellitus'].replace(to_replace = {' yes' : 'yes', '\tyes' : 'yes', '\tno' : 'no' }, inplace = True)
df['coronary_artery_disease'] = df['coronary_artery_disease'].replace(to_replace = '\tno', value = 'no')
df['class'] = df['class'].replace(to_replace = 'ckd\t', value = 'ckd')
```

```python
#Data Encoding (change value to 1 and 0) in class
```

```python
df['class'] = df['class'].map({'ckd' : 0, 'notckd' : 1})
```

```python
for col in cat_cols:
  print(f"{col} has {df[col].unique()} values \n")
```

#**Bivariate Analysis**

```python
df.columns
```

```python
#Create a scatterplot for Age and Blood Pressure
#As age increase Bp also increase.
sns.scatterplot(x = 'age', y = 'blood_pressure', data = df, hue = 'class', palette = 'dark')
plt.title('Scatter Plot for Age Vs Blood Pressure')
plt.xlabel('Age')
plt.ylabel('Blood Pressure')
plt.show()
```

**Insight**-Blood pressure values mostly range between 60 and 120, with a few outliers reaching up to 180. The plot shows that blood pressure tends to vary more widely in middle-aged individuals.

```python
#Create a Boxplot for Diabetes Mellitus and Albumin
sns.boxplot(x = 'diabetes_mellitus', y = 'albumin', data = df, palette = 'viridis' )
plt.title('Box Plot for Diabetes Mellitus and Albumin')
plt.xlabel('Diabetes Mellitus')
plt.ylabel('Albumin')
plt.show()
```

**Insight**-The median Albumin level is higher in individuals with Diabetes Mellitus compared to those without it. There are a couple of outliers in the non-diabetic group, showing higher than usual Albumin levels.

```python
#Create a Violineplot for Diabetes Mellitus and Albumin
sns.violinplot(x = 'diabetes_mellitus', y = 'albumin', data = df, palette = 'dark' )
plt.title('Violin Plot for Diabetes Mellitus and Albumin')
plt.xlabel('Diabetes Mellitus')
plt.ylabel('Albumin')
plt.show()
```

**Insight**-Diabetes Mellitus (Yes): The distribution is wider, indicating a greater variability in Albumin levels. The density peaks around 3, suggesting this is the most common value.
Diabetes Mellitus (No): The distribution is narrower and more uniform, with fewer high values and a peak around 1.

The median (white dot) is higher for those with Diabetes Mellitus, consistent with the box plot observation.
Variability:

The spread of Albumin values is wider in the diabetic group, indicating more variability. This is also shown by the presence of tails extending towards higher values.

```python
#Create a Stacked Bar Chart for Diabetes and Hypertension
diabetes_hypertension = pd.crosstab(df['diabetes_mellitus'], df['hypertension'])
diabetes_hypertension.plot(kind = 'bar', stacked = True)
plt.title('Stacked Bar Chart for Diabetes Mellitus and Hypertension')
plt.xlabel('Diabetes Mellitus')
plt.ylabel('Count')
plt.show()
```

**Insight**-The count of individuals without Diabetes Mellitus is higher than those with it. Among individuals without Diabetes Mellitus, those without hypertension form the majority. Among individuals with Diabetes Mellitus, there is a substantial count of individuals with hypertension.
